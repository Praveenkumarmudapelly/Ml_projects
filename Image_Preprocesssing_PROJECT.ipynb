### Function to take a color image :
### Do preprocessing steps:

* creat a PIL object
* Resize
* Color Enhance and change to Grey scale
* Create Numpy array of Grey Scale image
* Flatten the Numpy array and return it

```python
# Function to take one image and convert it in to flattened grey scale numpy array
def img_to_ary(path):
    imaPil = Image.open(path)
    imgResized = imaPil.resize((40,40))
    imgColE = ImageEnhance.Color(imgResized)
    imdNp = np.array(imgColE.enhance(0))
    img_flatten = imdNp.flatten()
    return img_flatten
```

### Create a target column


* donkey images will have text 'donkey' in it
* horses images will have text 'horses' in it
* Text will added based on what folder they are inputed from

#### Taking the Images from separate floder 

```python
def img_to_ary(path,name):
    imaPil = Image.open(path)
    imgResized = imaPil.resize((40,40))
    imgColE = ImageEnhance.Color(imgResized)
    imdNp = np.array(imgColE.enhance(0))
    img_flatten = imdNp.flatten()
    return list(np.append(img_flatten,name))
```

```python
import os
from os import listdir
# get the path/directory
```

```python


img_animal = ['horses','donkey']
preprocess = []

for j in img_animal:
    folder_dir = (r"C:\Users\prave\Downloads\imges\{}".format(j))
    
    for i in os.listdir(folder_dir):
        image_folder = (f"{folder_dir}\{i}")
        preprocess.append(img_to_ary(image_folder,j))

df = pd.DataFrame(preprocess,dtype=object)
df
```

```python
df.rename(columns = {4800:'Image_label'}, inplace = True)
df
```

```python
y=df["Image_label"]
X=df.drop(columns="Image_label")
```

### Spliting the data to Train and Test

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,
                                        test_size=0.20,
                                        random_state=23)
```

### Creating a Decision Tree model

```python
DT=DecisionTreeClassifier() # object

DT.fit(X_train,y_train)
```

```python
y_train_pred,y_test_pred=DT.predict(X_train),DT.predict(X_test)
```
